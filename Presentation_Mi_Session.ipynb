```python
Review_title =  []
Review = []
Date_of_visit = []

for i in (range(10, 50,10)):
    url = f'https://www.tripadvisor.co.uk/Restaurant_Review-g186338-d12156905-Reviews-or{i}-Bokan_37_Restaurant-London_England.html'
    html = requests.get(url).content
    response = Selector(text=html)
    
    Review_title.append(response.css('span.noQuotes::text').extract())
    Review.append(response.css('div.entry p::text').extract())
    Date_of_visit.append(response.css("div.prw_reviews_stay_date_hsx::text").extract())
    
review_item = {}
review_item['Review_title'] = [item for sublist in Review_title for item in sublist]
review_item['Review'] = [item for sublist in Review for item in sublist]
review_item["Date_of_visit"] = [item for sublist in Date_of_visit for item in sublist]
    
```
