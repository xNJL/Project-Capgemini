## 1. Make necessary imports

```python
import requests
from scrapy.selector import Selector
```

## 2. Get some information from Capgemini Invent's website

In order to discover what selectors are, you will extract some basic information from [this page](https://www.capgemini.com/service/invent/)

```python
# Define  the url
url = 'https://www.capgemini.com/service/invent/'
```

```python
# Create selector object
html = requests.get(url).content
sel = Selector(text=html)
```

## 3. Let's start with a basic example : extracting Capgemini Invent's moto with selectors

_As a reminder, you can use `F12` to inspect a webpage from your favorite navigator_

```python
# Getting the Capgemini Invent moto thanks to css selector
css_locator = 'h1 ::text'
rep = sel.css(css_locator).extract_first()
print(rep)
```

```python
# Doing the same with xpath selector
xpath_locator = '//h1//text()'
rep = sel.xpath(xpath_locator).extract_first()
print(rep)
```

__<font color='Blue'>
Exercice 1 : Using selectors, get the name of all the Capgemini Invent experts

```python
css_selector = 'h3.card_person__title ::text'
rep = sel.css(css_selector).extract()
rep = [i.replace('\t', '').replace('\n', '') for i in rep]
rep = [i for i in rep if len(i)>0]
list(set(rep))
```

__<font color='Blue'>
Exercice 5 : Using selectors get the name of at least 30 Capgemini client use-case
</font>__

```python
# Hint : you might have to find another page(s)

# Select url
url = 'https://www.capgemini.com/all-client-stories/?show_posts=30'

# Create selector object
html = requests.get(url).content
sel = Selector(text=html)

# Get
css_selector = 'h3.card_default__title ::text'
use_case = sel.css(css_selector).extract()
use_case = [i.replace('\t', '').replace('\n', '') for i in use_case]
use_case = [i for i in use_case if len(i)>0]

# Display result
print(len(use_case))
_ = [print(cas) for cas in use_case[0:10]]
```

__<font color='Blue'>
Exercice 6 : Now get all of them ;) !
</font>__

```python
# Hint : you might have to find another page(s)
urls = [
    'https://www.capgemini.com/all-client-stories/?show_posts=120',
    'https://www.capgemini.com/all-client-stories/page/2/?show_posts=120',
    'https://www.capgemini.com/all-client-stories/page/3/?show_posts=120',
    'https://www.capgemini.com/all-client-stories/page/4/?show_posts=120'
]

cas_client = []     
for url in urls:
    # Create selector object
    html = requests.get(url).content
    sel = Selector(text=html)
    css_selector = 'h3.card_default__title ::text'
    scrap = sel.css(css_selector).extract()
    scrap = [i.replace('\t', '').replace('\n', '') for i in scrap]
    scrap = [i for i in scrap if len(i)>0]
    cas_client += scrap
```

```python
print(len(cas_client))
_ = [print(cas) for cas in cas_client[0:10]]
```
